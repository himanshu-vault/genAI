```python
from langchain_core.tools import tool
```

#### steps to create a tool - 3 steps

```python
# create a function

def multiply(a,b):
    """multiply two numbers"""  #### mandatory docstring to let llm understand what is the function about
    return(a*b)
```

```python
# add type hinting

def multiply(a: int,b: int) -> int:
    """multiply two numbers"""  #### mandatory docstring to let llm understand what is the function about
    return(a*b)
```

```python
# add decorator to make it a tool
@tool
def multiply(a: int,b: int) -> int:
    """multiply two numbers"""  #### mandatory docstring to let llm understand what is the function about
    return(a*b)
```

```python
print("multiply.name          -   " ,  multiply.name)
print("multiply.description   -   " ,  multiply.description)
print("multiply.args          -   " ,  multiply.args)
```

#### how to use

```python
result = multiply.invoke({"a":3, "b":5})
print(result)
```
