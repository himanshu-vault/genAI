```python
from langchain_core.tools import BaseTool
from pydantic import BaseModel, Field
from typing import Type
```

```python
# ceating input structure
class multiply_input(BaseModel):
    a: int = Field(required=True, description='First number to multiply')
    b: int = Field(required=True, description='Second number to multiply')
```

```python
class MultiplyTool(BaseTool):
    name: str = "Multiply"
    description: str = "Multiply 2 numbers"
    
    args_schema: Type[BaseModel] = multiply_input

    def _run(self, a,b):
        """multiply two numbers"""  #### mandatory docstring to let llm understand what is the function about
        return(a*b)
```

```python
# instance of the class
multiply_tool = MultiplyTool()
```

```python
print("multiply_tool.name          -   " ,  multiply_tool.name)
print("multiply_tool.description   -   " ,  multiply_tool.description)
print("multiply_tool.args          -   " ,  multiply_tool.args)
```

#### how to use

```python
result = multiply_tool.invoke({"a":3, "b":5})
print(result)
```
