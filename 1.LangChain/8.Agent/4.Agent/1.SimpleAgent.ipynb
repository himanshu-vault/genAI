```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
import requests
from langchain_community.tools import DuckDuckGoSearchRun
```

```python
# agent specific 
# from langchain.agents import create_react_agent
from langchain_classic.agents import create_react_agent, AgentExecutor
from langchain_classic import hub

hwchase_prompt = hub.pull('hwchase17/react') # pulls the standard ReAct(REasoning + ACTing) agent prompt
hwchase_prompt
```

```python
# tool
search_tool = DuckDuckGoSearchRun()
result = search_tool.invoke('Top news in India')
```

```python
# llm
llm = ChatOpenAI()
```

```python
# create an agent - agent "thinks"
agent = create_react_agent(
    llm=llm,
    tools=[search_tool],
    prompt=hwchase_prompt
)
```

```python
# wrap it in agent executor - executor "executes"
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool],
    verbose=True
)
```

```python
agent_executor.invoke({'input' : "5 pointers on ongoing T20 matches"})
```
