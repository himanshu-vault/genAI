```python
from langchain_core.tools import BaseTool, tool, StructuredTool
from pydantic import BaseModel, Field
from typing import Type
from langchain_community.tools import DuckDuckGoSearchRun
```

```python
# search_tool = DuckDuckGoSearchRun()
```

```python
# decorator tool
@tool
def multiply(a: int,b: int) -> int:
    """multiply two numbers"""  #### mandatory docstring to let llm understand what is the function about
    return(a*b)
```

```python
# pydantic tool
class multiply_input(BaseModel):
    a: int = Field(required=True, description='First number to multiply')
    b: int = Field(required=True, description='Second number to multiply')

def multiply(a: int,b: int) -> int:
    """multiply two numbers"""  #### mandatory docstring to let llm understand what is the function about
    return(a*b)

multiply_tool = StructuredTool.from_function(
    func=multiply,
    name='multiply',
    description="Multiply 2 numbers",
    args_schema=multiply_input
)
```

```python
# BaseTool inheritance tool

# pydantic tool
class AddInput(BaseModel):
    a: int = Field(required=True, description='First number to Add')
    b: int = Field(required=True, description='Second number to Add')

class AddTool(BaseTool):
    name: str = "Add"
    description: str = "Add 2 numbers"
    
    args_schema: Type[BaseModel] = AddInput

    def _run(self, a,b):
        """Add two numbers"""  #### mandatory docstring to let llm understand what is the function about
        return(a+b)
    
# instance of the class
Add_tool = AddTool()
```

```python
class MathToolKit():
    def get_tools(self):
        return [Add_tool, multiply_tool]
```

#### how to use

```python
toolkit = MathToolKit()
tools = toolkit.get_tools()
for tool in tools:
    print("tool.name          -   " ,  tool.name)
    print("tool.description   -   " ,  tool.description)
    print("tool.args          -   " ,  tool.args, '\n')
```
