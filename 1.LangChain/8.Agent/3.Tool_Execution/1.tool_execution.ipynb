```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """given 2 numbers a and b this tool returns their product"""
    return a*b
```

```python
print("multiply.name          -   " ,  multiply.name)
print("multiply.description   -   " ,  multiply.description)
print("multiply.args          -   " ,  multiply.args, '\n')
```

```python
# this is direct invoke of a tool
multiply.invoke({"a":1, "b":2})
```

```python
# need to call the tool from LLM

model = ChatOpenAI()
model_with_tool = model.bind_tools([multiply])
```

```python
result = model_with_tool.invoke('Whats product of 8 and 3')
```

```python
# WTF?
result
```

```python
# LLM only identifies that the tool call is needed to be done
# it doesn't calls the function
# it has to be handled by you or langchain

result.tool_calls[0]
```

```python
args = result.tool_calls[0]['args']

# return only a response of the function converted to tool
multiply.invoke(args)
```

```python
# returns a ToolMessage like (human and system messages)
multiply.invoke(result.tool_calls[0])
```
