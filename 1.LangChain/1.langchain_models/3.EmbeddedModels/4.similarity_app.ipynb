```python
from langchain_openai import OpenAIEmbeddings
model = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=32)
```

```python
docs = [
    "Indian cricket has a rich history with legendary players like Sachin Tendulkar and Kapil Dev.",
    "The Indian cricket team won the ICC Cricket World Cup in 1983 and 2011.",
    "Cricket is the most popular sport in India and unites fans across the country.",
    "The Indian Premier League has transformed cricket into a global entertainment league.",
    "India is known for producing world-class batsmen, bowlers, and all-rounders."
]



query = "which is the most popular sport in India"
```

```python
docs_embedded = model.embed_documents(docs)
query_embedded = model.embed_query(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
scores = cosine_similarity([query_embedded], docs_embedded)
```
