```python
from dotenv import dotenv_values
env_file_vars = dotenv_values()
OPENAI_API_KEY = env_file_vars.get("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
model = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=32)
```

```python
docs = [
    "Delhi is the capital of India",
    "Kolkata is the capital of Bengal",
    "Paris is the capital of France"
]
```

```python
result = model.embed_documents(docs)
```
