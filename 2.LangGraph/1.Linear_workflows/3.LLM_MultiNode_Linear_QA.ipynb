```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# start > question to LLM > end
```

```python
model = ChatOpenAI()
```

```python
# creating state
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
# creating graph(liner)
graph = StateGraph(LLMState)
```

```python
def llm_qa(state: LLMState) -> LLMState:
    # extract question > form a prompt > ask LLM > update answer in state
    
    question = state['question']
    
    prompt = f"Answer the following question\n{question}"

    response = model.invoke(prompt).content

    state['answer'] = response

    return state


# adding nodes
graph.add_node('llm_qa', llm_qa)
```

```python
# adding edges
graph.add_edge(START, 'llm_qa')

graph.add_edge('llm_qa', END)
```

```python
# compile the graph
workflow = graph.compile()
```

```python
# execute the graph
final_state = workflow.invoke({"question":"Who was the creator of computer"})
final_state
```
