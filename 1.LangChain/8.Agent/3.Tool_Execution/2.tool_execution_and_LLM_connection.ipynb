```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """given 2 numbers a and b this tool returns their product"""
    return a*b
```

```python
# need to call the tool from LLM

model = ChatOpenAI()
model_with_tool = model.bind_tools([multiply])
```

```python
query = HumanMessage("Can you multiply 3 and 8000")
messages = [query]
messages
```

```python
result = model_with_tool.invoke(messages)
```

```python
# WTF?
messages.append(result)
```

```python
# returns a ToolMessage like (human and system messages)
tool_message = multiply.invoke(result.tool_calls[0])
```

```python
messages.append(tool_message)
```

```python
for i,message in enumerate(messages):
    print(i+1, '--------\n',message, '\n')
```

```python
final_result = model_with_tool.invoke(messages)
```
